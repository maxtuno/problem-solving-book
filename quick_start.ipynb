# Boolean Variables

```python
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
x = tc.int()
assert x <= 1
while tc.satisfy([x]):
    print(x)
```

# Two Boolean Variables

```python
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
assert x <= 1
assert y <= 1
while tc.satisfy([x, y]):
    print(x, y)
```

# Three Integer Variable 6 bits

```python
tc = peqnp.TheCore(bits=5, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
z = tc.int()
assert x + y == z
while tc.satisfy([x, y, z]):
    print(x, y, z)
```

# Three Integer Variable 4 bits Operations

```python
tc = peqnp.TheCore(bits=4, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
z = tc.int()
assert x ** 2 + y ** 3 == x * y - z
while tc.satisfy([x, y, z]):
    print(x, y, z)
```

# Three Integer Variable 10 bits Exponential Variables

```python
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
z = tc.int()
assert x ** y == z
assert x > 1
assert y > 1
while tc.satisfy([x, y, z]):
    print(x, y, z)
```

# Three Integer Variable 10 bits Factorial

```python
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
assert tc.factorial(x) + 1 == y ** 2
while tc.satisfy([x, y]):
    print(x, y)
```

# Array of 4 Boolean Variables Constraints

```python
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
xs = tc.array(4)
assert sum(xs) == 1
while tc.satisfy(xs): # note array is passed directly
    print(tc.values(xs))
```

# Array of 4 Integers "All Differents"

```python
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
xs = tc.array(4)
tc.apply(xs, single=lambda x: x < 5) # all values < 5
tc.apply(xs, dual=lambda x, y: x != y)
while tc.satisfy(xs): # note array is passed directly
    print(tc.values(xs))
```

# Matrix of 4x4 Boolean Variables Constraints

```python
tc = peqnp.TheCore(bits=4, key='tutorial', folder='db/')
xs = [tc.array(2) for _ in range(2)]
for i in range(2):
    assert sum(xs[i]) == 1
while tc.satisfy(tc.flatten(xs)): # note matrix is flatten to array and passed directly
    print(tc.values(tc.flatten(xs)))
    for i in range(2):
        print(tc.values(xs[i]))
```

```python
import numpy
tc = peqnp.TheCore(bits=10, key='tutorial', folder='db/')
xs = numpy.array([tc.array(4) for _ in range(4)])
for i in range(4):
    assert sum(xs[i]) == 1
    assert sum(xs.T[i]) == 1
while tc.satisfy(tc.flatten(xs)): # note matrix is flatten to array and passed directly
    print(tc.values(tc.flatten(xs)))
    for i in range(4):
        print(tc.values(xs[i]))
```

# Matrix Operations With Numpy

```python
import numpy
tc = peqnp.TheCore(bits=4, key='tutorial', folder='db/')
xs = numpy.reshape(numpy.array([tc.array(2) for _ in range(2)]), (2, 2))
ys = numpy.reshape(numpy.array([tc.array(2) for _ in range(2)]), (2, 2))
for i in range(2):
    assert sum(xs[i]) == 1
    assert sum(xs.T[i]) == 1
    for j in range(2):
        assert ys[i][j] <= 1
assert (numpy.multiply(xs, ys) <= 1).all()
while tc.satisfy(tc.flatten(xs) + tc.flatten(ys)): # note matrix is flatten to array and passed directly
    print(numpy.vectorize(int)(xs))
    print(numpy.vectorize(int)(ys))
    print(numpy.multiply(numpy.vectorize(int)(xs), numpy.vectorize(int)(ys))) # check
    print()
```

# Exponential Diophantine Equations

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
z = tc.int()
_3 = tc.int(value=3)
assert _3 ** x == y ** 5 + z
assert x > 1
assert y > 1
while tc.satisfy([x, y, z]):
    print(x, y, z)
```

# Permutations

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
z = tc.int()
tc.permutations([1, 2, 3], [x, y, z])
while tc.satisfy([x, y, z]):
    print(x, y, z)
```

# Subsets

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
xs = tc.subset(3, [1, 2, 3, 4, 5], 0) # a subset of {1, 2, 3, 4, 5} of at most 3 elements and empty element is 0
while tc.satisfy(xs):
    print(tc.values(xs))
```

# Sequencing

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
xs = tc.array(3)
ys = tc.array(3)
tc.sequencing(xs, ys, [1, 2, 3, 4]) # xs positions, ys elements from array
while tc.satisfy(xs + ys):
    print(tc.values(xs), tc.values(ys))
```

# IFF

```python
tc = peqnp.TheCore(bits=4, key='tutorial', folder='db/')
x = tc.int()
bit = tc.int(size=1)
assert x.iff(bit[0], tc.zero) == tc.zero.iff(-bit[0], x) 
while tc.satisfy([x, bit]):
    print(bit.binary, x)
```

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
x = tc.int()
y = tc.int()
z = tc.int()
bit = tc.int(size=1)
assert (x ** 3 + y ** 3 - z ** 3).iff(bit[0], x ** 3 - y ** 3 - z ** 3) == 10
while tc.satisfy([x, y, z, bit]):
    if not bit.binary[0]:
        print(x ** 3 + y ** 3 - z ** 3)
    else:
        print(x ** 3 - y ** 3 - z ** 3)
```

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
data = [2, 3, 5, 7, 11]
bit = tc.int(size=5)
assert sum(tc.zero.iff(bit[i], data[i]) for i in range(5)) == 12
while tc.satisfy([bit]):
    print(bit.binary)
```

```python
tc = peqnp.TheCore(bits=16, key='tutorial', folder='db/')
data = [2, 3, 5, 7, 11]
xs = tc.subset(5, data, tc.zero)
assert sum(xs) == 12
while tc.satisfy(xs):
    print(tc.values(xs))
```
